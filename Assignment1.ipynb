**Variable 2**

```python
# Variable 2
# Fuel Type : fuel type with fuelType1 and fuelType2 (if applicable)

# Counting the number of fuelType for each year and category
fuel_counts = df.groupby(['year', 'fuelType']).size().unstack(fill_value=0)
# print(car_counts)

# Generate a line plot for each category in atvType
for category in fuel_counts.columns:
    plt.plot(fuel_counts.index, fuel_counts[category], label=category)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Fuel Type Vehicles')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```

```python
# Fuel Type 1
# fuelType1 : fuel type 1. For single fuel vehicles, this will be the only fuel. For dual fuel vehicles, this will be the conventional fuel.

# Counting the number of fuelType1 for each year and category
fuel_counts = df.groupby(['year', 'fuelType1']).size().unstack(fill_value=0)
# print(car_counts)

# Generate a line plot for each category in atvType
for category in fuel_counts.columns:
    plt.plot(fuel_counts.index, fuel_counts[category], label=category)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('fuelType1')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```

```python
# Fuel Type 2
# fuelType2 : fuel type 2. For dual fuel vehicles, this will be the alternative fuel (e.g. E85, Electricity, CNG, LPG). For single fuel vehicles, this field is not used

# Counting the number of fuelType2 for each year and category
fuel_counts = df.groupby(['year', 'fuelType2']).size().unstack(fill_value=0)
# print(car_counts)

markers = ['o', 's', '^', 'D', 'v']

# Generate a line plot for each category in atvType
for i, category in enumerate(fuel_counts.columns):
    plt.plot(fuel_counts.index, fuel_counts[category],label=category, marker=markers[i])

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Fuel Type 2')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```
